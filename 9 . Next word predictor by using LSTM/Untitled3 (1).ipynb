```python
import pandas as pd
import numpy as np
with open("shakespeare_gutenberg.txt" , "r" , encoding='utf-8') as file:
    data = file.read() 
    data = data[:5000]
    print(data)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
# Tokenizing our data
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
vocab_size = len(tokenizer.word_index) + 1 
print(vocab_size)
```

```python
input_sequence = []
for line in data.split("\n"):
    token_list =  tokenizer.texts_to_sequences([line])[0]
    for i in range(1 , len(token_list)):
        n_gram_sequences = token_list[:i+1]
        input_sequence.append(n_gram_sequences)
max_len = max(len(seq) for seq in input_sequence )
print(max_len)
input_sequence
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
pad_sequences_data = np.array(pad_sequences(input_sequence , maxlen = max_len ))
pad_sequences_data
```

```python
x = pad_sequences_data[:,:-1]
y = pad_sequences_data[:,-1]
y = tf.keras.utils.to_categorical(y , num_classes=vocab_size)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y , random_state=2 , test_size=0.2) 
```

```python
print(x_train.shape , x_test.shape , 
y_train.shape , y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM , Dense , Embedding  ,Dropout ,GRU
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size ,input_length = max_len, output_dim=8))
model.add(LSTM(50 , return_sequences=True ))
model.add(LSTM(50))
model.add(Dense(vocab_size , activation="softmax"))
```

```python
model.compile(loss = "categorical_crossentropy" , optimizer="adam" , metrics=["accuracy"])
model.fit(x,y , batch_size=32 , epochs=200)
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size ,input_length = max_len, output_dim=8))
model.add(GRU(50 , return_sequences=True ))
model.add(GRU(50))
model.add(Dense(vocab_size , activation="softmax"))
```

```python
model.compile(loss = "categorical_crossentropy" , optimizer="adam" , metrics=["accuracy"])
model.fit(x,y , batch_size=32 , epochs=600 ,verbose='auto')
```

```python
import time
for i in range(max_len):
    input_data = "THE FIRST PART OF KING"
# Tokenize
    tokenized_data = tokenizer.texts_to_sequences([input_data])[0]
# padding
    padd_data = pad_sequences([tokenized_data] , maxlen=max_len  )
#predict
    predic_data = np.argmax(model.predict(padd_data))


for word,index in tokenizer.word_index.items():
    if index == predic_data:
      text = input_data + " " + word
      print(text)
      time.sleep(2)
```
